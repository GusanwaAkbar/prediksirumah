```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
ratios = ["90:10", "80:20", "70:30", "60:40", "50:50"]
mae_values = [83681975.0, 120314881.25, 106286529, 156089548, 168542525]
mape_values = [15.10, 13.17, 11.29, 14.64, 16.09]

# Plot MAE
plt.figure(figsize=(10, 5))
sns.barplot(x=ratios, y=mae_values, color='blue')
plt.title('Mean Absolute Error (MAE) for Different Ratios')
plt.xlabel('Ratio of Data Split (Train:Test)')
plt.ylabel('MAE Value')
plt.show()

# Plot MAPE
plt.figure(figsize=(10, 5))
sns.barplot(x=ratios, y=mape_values, color='orange')
plt.title('Mean Absolute Percentage Error (MAPE) for Different Ratios')
plt.xlabel('Ratio of Data Split (Train:Test)')
plt.ylabel('MAPE (%)')
plt.show()
```
